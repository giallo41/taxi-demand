## Multi-Task 비교 (SEO vs NYC )

#### 1) Base ( Start (t0) + End (t0) + Temp (t1)  ) -> Y (t1) 

#### 2) Multi 2 task ( Start (t0) + End (t0) + Temp (t1)  ) -> Y (t1), Y(t2) - t1, t2 shared embedding 

#### 3) Base ( Start (t0) + End (t0) + Temp (t2)  ) -> Y (t2) 

```python
print ('2) Base : t1 & t2 embedding -> t1 => Y(t1) & t2=>Y(t2) ')
DATA_FOLDER = '/data/public/rw/prj-mobility/taxi-demand/output_file/MODEL_V1_SEO/'
m_name = 'MODEL_V1_SEO_0827_005_SEO_V1_base_3b'

def chk_2ch_output(model_name):
    
    df_t1_pred = pd.read_csv(model_name+'t1_pred.csv')
    df_t1_true = pd.read_csv(model_name+'t1_gt.csv')
    
    df_t2_pred = pd.read_csv(model_name+'t2_pred.csv')
    df_t2_true = pd.read_csv(model_name+'t2_gt.csv')
    
    df_t1_true = df_t1_true.drop('Unnamed: 0', axis=1)
    df_t1_pred = df_t1_pred.drop('Unnamed: 0', axis=1)
    
    df_t2_true = df_t2_true.drop('Unnamed: 0', axis=1)
    df_t2_pred = df_t2_pred.drop('Unnamed: 0', axis=1)
    
    
    
    y_t1_pred = df_t1_pred.values
    y_t1_true = df_t1_true.values
    
    y_t2_pred = df_t2_pred.values
    y_t2_true = df_t2_true.values
    
    
    y_t1_pred = np.expand_dims(y_t1_pred, axis=2)
    y_t1_true = np.expand_dims(y_t1_true, axis=2)
    
    y_t2_pred = np.expand_dims(y_t2_pred, axis=2)
    y_t2_true = np.expand_dims(y_t2_true, axis=2)
    
    
    
    y_pred = np.concatenate((y_t1_pred, y_t2_pred), axis=-1)
    y_true = np.concatenate((y_t1_true, y_t2_true), axis=-1)
    
    print ('MAPE 11 : %.4f'%(mape_trs(y_true, y_pred, 11)))
    print ('RMSE 11 : %.4f'%(rmse_trs(y_true, y_pred, 11)))
    print ('MAE 11 : %.4f'%(mae_trs(y_true, y_pred, 11)))
    print ('')
    print ('MAPE all : %.4f'%(mape(y_true, y_pred)))
    print ('RMSE all : %.4f'%(rmse(y_true, y_pred)))
    print ('MAE all : %.4f'%(mae(y_true, y_pred)))
    print ('')
    print ('T1 MAPE 11 : %.4f'%(mape_trs(y_t1_true, y_t1_pred, 11)))
    print ('T2 MAPE 11 : %.4f'%(mape_trs(y_t2_true, y_t2_pred, 11)))
    print ('')
    print ('T1 RMSE 11 : %.4f'%(rmse_trs(y_t1_true, y_t1_pred, 11)))
    print ('T2 RMSE 11 : %.4f'%(rmse_trs(y_t2_true, y_t2_pred, 11)))
    print ('')
    print ('T1 MAE 11 : %.4f'%(mae_trs(y_t2_true, y_t2_pred, 11)))
    print ('T2 MAE 11 : %.4f'%(mae_trs(y_t2_true, y_t2_pred, 11)))
    print ('')
    print ('T1 MAPE all : %.4f'%(mape(y_t1_true, y_t1_pred)))
    print ('T2 MAPE all : %.4f'%(mape(y_t2_true, y_t2_pred)))
    print ('')
    print ('T1 RMSE all : %.4f'%(rmse(y_t1_true, y_t1_pred)))
    print ('T2 RMSE all : %.4f'%(rmse(y_t2_true, y_t2_pred)))
    print ('')
    print ('T1 MAE all : %.4f'%(mae(y_t2_true, y_t2_pred)))
    print ('T2 MAE all : %.4f'%(mae(y_t2_true, y_t2_pred)))
    
chk_2ch_output(DATA_FOLDER+m_name)
```
