## Data Stats / Temporal Feature

```python
DATA_FOLDER = '/data/public/rw/prj-mobility/10.195.12.143/kakaobrain/taxi/datasets/model_2018/'
x_st_train = np.load(DATA_FOLDER+'X_train_data_st_30min_16ch_ori.npz')['arr_0'][:,:,:,:1]
x_ed_train = np.load(DATA_FOLDER+'X_train_data_ed_30min_16ch_ori.npz')['arr_0'][:,:,:,:1]
time_train = np.load(DATA_FOLDER+'time_train_data_30min_16ch_ori.npz')['arr_0']
week_train = np.load(DATA_FOLDER+'week_train_data_30min_16ch_ori.npz')['arr_0']
holi_train = np.load(DATA_FOLDER+'holi_train_data_30min_16ch_ori.npz')['arr_0']
prev_train = np.load(DATA_FOLDER+'prev_train_data_30min_16ch_ori.npz')['arr_0']

np_x_st_train = np.reshape(x_st_train, (5744, 2500))
np_x_ed_train = np.reshape(x_ed_train, (5744, 2500))

x_st_train.shape, x_ed_train.shape, time_train.shape, week_train.shape, holi_train.shape, prev_train.shape
```

```python
def apply_weekend(col):
    if col == 0 | col == 6:
        return 1
    else:
        return 0
    
```

```python
col_name = []
for i in range(0, 2500):
    col_name.append('col_'+str(i) )

df_st_train = pd.DataFrame(np_x_st_train, columns=col_name, index=np.arange(0,np_x_st_train.shape[0]))
#df_st_train = pd.DataFrame()
df_st_train['week'] = np.reshape(week_train, (5744,))
df_st_train['time'] = np.reshape(time_train, (5744,))
df_st_train['holi'] = np.reshape(holi_train, (5744,))
df_st_train['prev'] = np.reshape(prev_train, (5744,))
df_st_train['weekend'] = df_st_train['week'].apply(apply_weekend)

df_st_train.head()
```

```python
# Ref - http://data.seoul.go.kr/dataList/datasetView.do?infId=10790&srvType=A&serviceKind=2
#거주지 - 은평구, 노원구, 송파구 강서구, 관악구 
#상업지구 - 종로구, 중구, 강남구, 영등포구

office_col = [1075, 1078, 1175, 1176, 1225, 1226, 1527, 1528, 1568, 1618, 1619, 1632, 1631, 1633, 1669, 1680, 1681, 1682, 1784, 1831, 1832, 1881, 1882]
resid_col = [579, 629, 630, 631, 632, 683, 686, 731, 778, 779, 780, 830, 867, 868, 1017, 1018, 1167, 1168, 1207, 1236, 1510, 1511, 1771, 1772, 1821, 1822, 1823, 1829, 1838, 1839, 1840, 1921, 1922, 1927, 1926, 1945, 1971, 1972, 1973, 1976, 1977, 1978, 1979, 1980]

office_col_name = []
resid_col_name = []

for itm in office_col:
    office_col_name.append('col_'+str(itm) )

for itm in resid_col:
    resid_col_name.append('col_'+str(itm) )

```

```python
office_tm = []
resid_tm = []

for tm in range(0, 48):
    df_selected = df_st_train[(df_st_train['time']==tm)].iloc[:,:2500]
    df_tmp_office = df_selected[office_col_name]
    df_tmp_resid = df_selected[resid_col_name]
    
    office_tm.append(np.average(df_tmp_office.values))
    resid_tm.append(np.average(df_tmp_resid.values))

office_tm = np.asarray(office_tm)
resid_tm = np.asarray(resid_tm)

office_tm = office_tm/office_tm.max()
resid_tm = resid_tm/resid_tm.max()
```
